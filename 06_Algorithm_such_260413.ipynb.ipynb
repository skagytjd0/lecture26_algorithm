```python
def binary_search_recursive(A, key, low, high) :
    if (low> high) :
        return -1 #찾는 key가 A에 없는 경우 <- 탐색할 항목이 더 이상 없음

    middle = (low + high) //2 #중앙값 찾기

    # 중앙에 있는 값과 비교하여 같으면 해당 index를 return
    if (key == A[middle]):
        return middle
    elif (key < A[middle]) :# 담색 영역을 왼쪽으로 설정
        return binary_search_recursive(A, key, low, middle - 1)
    else : # 담색 영역을 오른쪽으로 설정
        return binary_search_recursive(A, key, middle + 1, high)
```

```python
def binary_search(A, key):
    low = 0
    high = len(A) - 1
    while low <= high:
        middle = (low + high) // 2
        if A[middle] == key:
            return middle
        elif A[middle] > key:
            high = middle - 1
        else:
            low = middle + 1
    return -1
```

```python
data = [2, 6, 11, 13, 18, 20, 22, 27, 29, 30, 34, 38, 41,42,45,47]
num = int(input('찾을 숫자 :'))

index = binary_search_recursive(data, num, 0, len(data) - 1)
if index == -1:
    print('없는 숫자입니다.')
else:
    print(f'>>> 숫자 {num}은 index {index}에 있습니다.')

index = binary_search(data, num)
if index == -1:
    print('없는 숫자입니다.')
else:
    print(f'>>> 숫자 {num}은 index {index}에 있습니다.')
```

```python
test_data = [34, 2, 47, 15, 48, 0]
answer = [10, 0, 15, -1, -1, -1]

for i in range(6):
    if binary_search(data, test_data[i]) == answer[i]:
        print('Pass : ', test_data[i])
    else:
        print('Fail : ', test_data[i])
```
