```python
stack = []
capacity = 5

def isEmpty():
    return len(stack) == 0

def isFull():
    return len(stack) == capacity

def push(e):
    if isFull():
        print("> Stack이 가득 차 있어서 더 이상 추가할 수 없습니다.")
    else:
        stack.append(e)

def pop():
    if isEmpty():
        print("> Stack이 비어 있습니다.")
        return None
    else:
        return stack.pop()

def peek():
    if isEmpty():
        print("> Stack이 비어 있습니다.")
        return None
    else:
        return stack[-1]

print(f"[[ 정수형 스택 연산 실습 (용량 : {capacity}) ]]")

while True:
    print("\n1. Push  2. Pop  3. Peek  0. Exit")
    choice = input("> 메뉴 선택: ")
    
    if not choice.strip():
        continue
        
    menu = int(choice)

    if menu == 0:
        break
        
    elif menu == 1:
        data = int(input("데이터 입력: "))
        push(data)
        
    elif menu == 2:
        result = pop()
        if result is not None:
            print(f"> 가져온 데이터 : {result}")
            
    elif menu == 3:
        result = peek()
        if result is not None:
            print(f"> 가져올 데이터 : {result}")
    
    print(f"> 현재 스택 상태 {stack}")

print("[[ 정수형 스택 연산 실습 종료 ]]")
```
