```python
class TreeNode:
    def __init__(self,data,left,right):
        self.data = data
        self.left = left
        self.right = right

    def set_left(self,left):
        self.left = left
    def set_right(self,right):
        self.right = right
```

```python
my_tree = TreeNode('A', None, None)
new_node = TreeNode('B', None, None)
my_tree.set_right(new_node)
my_tree.set_left(TreeNode('C', None, None))
cur_node = my_tree.left
cur_node.set_left(TreeNode('D', None, None))
cur_node.set_right(TreeNode('E', None, None))
cur_node = my_tree.right
cur_node.set_left(TreeNode('F', None, None))
```
