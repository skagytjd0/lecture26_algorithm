```python
# selection sort
# 정렬이 안된 부분 (start 이후)에서 가장 작은 수를 찾아서
# start에 있는 수와 swap
# start += 1
def print_step(step, data):
    print(f'단계 {step+1} : {data}')


def select(A, start):
    min = A[start]
    min_index = start
    
    for i in range(start+1, len(A)):
        if A[i] < min:
            min = A[i]
            min_index = i
    
    return min_index
    
def selection_sort(data_list):
    for start in range(len(data_list)-1):
        min_index = select(data_list, start)
        data_list[start], data_list[min_index] = data_list[min_index], data_list[start]
        print_step(start, data_list)
        
A = [23, 5, 44, 1, 6]
print('원래 데이터 : ', A)
selection_sort(A)
print('정렬 후 데이터 : ', A)
```

```python
# insertion_sort
def insertion_sort(A):
    for cur_index in range(1, len(A)):
        # 삽입할 데이터보다 작거나 같은 데이터를 만날 때까지
        # 데이터를 한 칸씩 오른쪽으로 복사
        key = A[cur_index]
        i = cur_index -1
        while i >= 0 and A[i] > key:
            A[i+1] = A[i]
            i -= 1
        # 현재 자리에 cur_index의 데이터를 저장
        A[i+1] = key
        print_step(cur_index, A)
        
```

```python
# bubble sort
def bubble_sort(A):
    for last in range(len(A)-1, 0, -1):
        is_changed = False
        for i in range(last):
            if A[i] > A[i+1]:
                A[i], A[i+1] = A[i+1], A[i]
                is_changed = True
    print_step(last, A)
    if not is_changed: return
```

```python
A = [1, 23, 5, 14, 21, 44, 1, 6]
print('원래 데이터 : ', A)
#selection_sort(A)
#insertion_sort(A)
bubble_sort(A)
print('정렬 후 데이터 : ', A)
```
